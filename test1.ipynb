```python
import os
import shutil
import pandas as pd
pd = pd.read_csv('sorted_output_case5.csv')
files_to_copy = pd['file_name'].tolist()
label_to_copy = pd['label'].tolist()
# 目标文件夹
destination_folder = 'review_case5'

# 确保目标文件夹存在，如果不存在则创建
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# 复制文件
for i in range(len(files_to_copy)):
    file_path = files_to_copy[i]
    label = label_to_copy[i]
    if os.path.exists(file_path):

        shutil.copy(file_path, destination_folder)
        print(f"Copied {file_path} to {destination_folder}")
    else:
        print(f"File {file_path} does not exist")
```

```python
import pandas as pd

# 读取 CSV 文件
input_csv_path = 'output_good.csv'
output_csv_path = 'sorted_output_case5.csv'

# 读取数据到 DataFrame
df = pd.read_csv(input_csv_path)

# 提取 'file_name' 列最后一个 '_' 后的整数部分，并作为新列添加到 DataFrame 中
df['sort_key'] = df['file_name'].apply(lambda x: int(x.split('_')[-1].split('.')[0]))

# 按照新列 'sort_key' 进行排序
df_sorted = df.sort_values(by='sort_key')

# 删除排序辅助列 'sort_key'
df_sorted = df_sorted.drop(columns=['sort_key'])

# 将排序后的 DataFrame 写回新的 CSV 文件
df_sorted.to_csv(output_csv_path, index=False)

print(f"CSV 文件已按 'column_name' 排序并保存到 {output_csv_path}")
```

```python
import pandas as pd
import random
import csv
file_path = "data_label.csv"
df = pd.read_csv(file_path)
label = df['label'].tolist()
path = df['img_path'].tolist()
new_csv = "train_v2.csv"
good_case = []
bad_case = []
bad = 0
good = 0
for i in range(len(label)):
    if label[i] == 0:
        good_case.append([path[i],0])
        good += 1
    else:
        bad_case.append([path[i],1])
        bad += 1
random.shuffle(good_case)
sample = random.sample(good_case, bad)
case = sample + bad_case
print(len(case))
with open(new_csv, 'w', newline='') as f:
    writer = csv.writer(f)
    # 写入列名
    writer.writerow(['file_name', 'label'])
    # 写入 case 数据
    writer.writerows(case)
```
